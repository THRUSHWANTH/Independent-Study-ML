```python
import pandas as pd
import os
```

```python
directory = "C:\\Users\\KAKUT\\Desktop\\Independent study\\csv_file\\"
years = list(range(2000, 2021))  # data from 2000 to 2021

df_list = []

for year in years:
    # Adjust the filename formatting to match the example you provided
    filename = f"MERGED{year}_{str(year+1)[-2:]}_PP.csv"
    file_path = os.path.join(directory, filename)
    
    if os.path.exists(file_path):
        df = pd.read_csv(file_path,low_memory=False)
        
        # Add a 'YEAR' column to the dataframe to record the starting year of data
        df['YEAR'] = year
        df_list.append(df)
    else:
        print(f"File {filename} not found.")


merged_df = pd.concat(df_list, axis=0, ignore_index=True)
```

```python
merged_df.head()
```

```python
merged_df.to_csv(os.path.join(directory, 'Mostmerged_data.csv'), index=False)
```
